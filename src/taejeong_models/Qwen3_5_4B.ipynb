```python
from transformers import pipeline
import torch

pipe = pipeline(
    "text-generation",
    model="Qwen/Qwen3.5-4B",
    device_map="auto",
    torch_dtype=torch.bfloat16
)

prompt = "한국어로 자기소개를 두 문장으로 해줘."
out = pipe(prompt, max_new_tokens=100)

print(out[0]["generated_text"])
```
