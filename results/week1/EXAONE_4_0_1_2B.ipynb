```python
from transformers import pipeline

pipe = pipeline(
    "text-generation",
    model="LGAI-EXAONE/EXAONE-4.0-1.2B",
    trust_remote_code=True,
    torch_dtype="auto",
    device_map="auto",

)

messages = [
    {"role": "user", "content": "한문장으로 자기소개"},
]
pipe(messages)
```
